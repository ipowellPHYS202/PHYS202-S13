###Differentiation

```python
def finiteDifference(x,y):
    """Here I define the function to take the small slope of the line y2-y1/x2-x1 where 
          y2-y1 is diff(y) and is small. The dydx=zeros(*args) sets the dimmension to bet that of y so our
    derivative has the same dimmension"""
    dydx = zeros(y.shape,float)
    dydx[:-1] = diff(y)/diff(x)
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx

```

```python
def fourPtFiniteDiff(x,y):
    """Uses a finite difference method to find the derivative of y with respect to x"""
    dydx = zeros(y.shape,float) #make the dimmension appropriate
    for j in range(2,len(y)-2):
        dydx[j] = (y[j-2]-8*y[j-1]+8*y[j+1]-y[j+2])/(12*(x[1]-x[0]))
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    return dydx
```

```python
#Using some numerical differentiation woo:

x = linspace(0,2*pi,100)
y = sin(x) + 0.1*random.random(size=x.shape)

plot(x,y, label = "original data")
DIFF=fourPtFiniteDiff(x,y)
plot(x,DIFF, 'k--', linewidth = 2, label="4 point centered diff")

plot(x,cos(x), linewidth=2, label = 'analytical derivative')

plot(x,finiteDifference(x,y), label='2 point forward diff')

xlabel('x')
ylabel('y')
xlim(0,6.5)
ylim(-4,2.3)
legend(loc='lower left')
show()
```

#Monte Carlo Multiple integration

```python
l = 0.2
w = 0.2
h = 1.0
N = 10000000
V=(l*w*h)

rho = lambda y,z: 100*((y+.1)+z**2)
```

```python
x = l*random_sample(N)
y = w*random_sample(N)
z = h*random_sample(N)

Mass = (V/N)*sum(rho(y,z))
print Mass
#(THIS IS A TRIPLE INTEGRAL!)
```

```python
r = lambda x,y,z: sqrt((x**2)+(y**2)+(z**2))*rho(y,z)

r_cm = (V/N)*sum(r(x,y,z))
print r_cm
#TRIPLE INTEGRAL FTW
```

```python
I = lambda x,y,z: ((sqrt((x**2)+(y**2)+(z**2))-r_cm)**2)*rho(y,z)

I_cm = (V/N)*sum(I(x,y,z))
print I_cm
```
