```python
def f(x):
    return x**.5+sin(x)
```

```python
x= arange(0.,15.,.001)
```

```python
myplot=plot(x,f(x),lw=3)
ylabel(r'$x^{1/2}+\sin(x)$', fontsize=20)
xlabel(r'$x$',fontsize=20)
xlim(0.,15.)
```

```python
xmin = 1.7
xmax = 12.4

#find max of f within the boundary to set ymax
subx=x[logical_and((x>xmin),(x<xmax))]
ymax = max(f(subx))*1.05
ymin=0.

#Replot the function
myplot=plot(x,f(x),lw=3)
xlim(0.,15.)

plot([xmin,xmin],[ymin,ymax],'k--')
plot([xmax,xmax],[ymin,ymax],'k--')
plot([xmin,xmax],[ymax,ymax],'k--')
show()
```

```python
#random_sample?

N=100000000
samples_x=xmin+(xmax-xmin)*random_sample(N);
samples_y = ymin+(ymax-ymin)*random_sample(N);

#Replot the function
myplot=plot(x,f(x),lw=3)
xlim(0.,15.)
plot(samples_x[1:200],samples_y[1:200],'.')

#plot the box
plot([xmin,xmin],[ymin,ymax],'k--')
plot([xmax,xmax],[ymin,ymax],'k--')
plot([xmin,xmax],[ymax,ymax],'k--')
show()
```

```python
newmask=(samples_y<f(samples_x))
```

```python
sum(newmask)
```

```python
#Replot the function
myplot=plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],'.')

#Show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'r.')

plot([xmin,xmin],[ymin,ymax],'k--')
plot([xmax,xmax],[ymin,ymax],'k--')
plot([xmin,xmax],[ymax,ymax],'k--')
show()
```

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

```python
Iana = lambda x: (2./3.)*x**(3./2.)-cos(x)
Iexact = Iana(xmax)-Iana(xmin)
Ierr = abs(100*(Iexact-I)/Iexact)
print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
    
```

```python
f=lambda x: x**(1./2.)+sin(x)
x = arange(0,15,.001)
xmin = 1.7
xmax = 12.4

#find max of f within the boundary to set ymax
subx=x[logical_and((x>xmin),(x<xmax))]
ymax = max(f(subx))*1.05
ymin=0
```

```python
N=1000
x_i=xmin+(xmax-xmin)*random_sample(N);
y_i = ymin+(ymax-ymin)*random_sample(N);
```

```python
total = 0
count=1
for i in x_i:
    int_= (xmax-xmin)
    count+=count
    total+=(int_/count)*(f(i))
print total
    
```
