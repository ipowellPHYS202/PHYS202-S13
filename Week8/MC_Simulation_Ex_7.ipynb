The Weibull distribution, with parameters 


$\alpha$ > 0 and $\beta$


 > 0 is described by the density



function

$f(x) = {\alpha}{\beta}^{-\alpha}{x}^{\alpha-1}{e^{-(x/\beta)}}$ for x>0

${f(x) = 0}$ otherwise

```python
from numpy import random
```

```python
#using the exact same technology from the notes:
f = lambda x: a*(b**(-a))*(x**(a-1))*exp(-(x/b)**a)
CDF = lambda x: 1-exp(-(x/b)**(a))
Finv = lambda z: b*(-log(1-z))**(1/a)
#Finv was found analytically

a = 1.5
b = 6
#Just following procedure:
N = 10000
x = arange(0,15,0.001)
z = numpy.random.random_sample(N)

X_Samples = Finv(z)

hist(X_Samples,100,normed=True,label='samples')
plot(x,f(x),label='actual',lw=3,color='r')
plot(x,CDF(x),label='cumulative',lw=3,color='k')
xlabel('x')
ylabel('y')

xlim(0,15)
ylim(0,1)

legend(loc=0)
show()
```

```python
#Beta(4,3) Distribution
f = lambda x: 60*(x**3)*((1-x)**2) #0<x<1
    
#Using the same technology as our Monte Carlo Rejection Sampling for Integration:
    
N = 10000 #10,000 samples
#defining maxs and mins with given data:   
x_min=0.
x_max=1.
y_min=0.
y_max=2.0736

samples_x=x_min +(x_max-x_min)*random_sample(N)
samples_y=y_min+(y_max-y_min)*random_sample(N)
#creating the Boolean Mask
AMask = samples_y < f(samples_x)
#print AMask
x=arange(0.,1.001,0.001)


#CDF, Normalized

CDF=cumsum(f(x))/sum(f(x))
#plotting all
plot(x,f(x),lw = 3, label='true Distribution')
plot(x,CDF,lw=3,label='CDF')
hist(samples_x[AMask],100,normed=True,label='samples distribution')
legend(loc=0)
xlabel('x')
ylabel("y")
title("Distributions and CDF")
show()
```
