###Question 3

```python
def finiteDifference(x,y):
    """Here I define the function to take the small slope of the line y2-y1/x2-x1 where 
    y2-y1 is diff(y) and is small. The dydx=zeros(*args) sets the dimmension to bet that of y so our
    derivative has the same dimmension"""
    dydx = zeros(y.shape,float)
    dydx[:-1] = diff(y)/diff(x)
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

```python
x = arange(0,10.1,.1) 
y = x**5 
y_prime = 5*x**4 #the exact derivative
plot(x,finiteDifference(x,y)) #the numerical result plotted
plot(x,y_prime) #the exact result plotted
show() 
```

```python
x_new=arange(0,10.001,.001)
y_new = x_new**5
dydx_new = finiteDifference(x_new,y_new)
y_prime_new = 5*x_new**4
percent_error = abs(y_prime_new-dydx_new)/y_prime_new #absolute value to keep percent positive
```

```python
#Error at 0 approaches a huge number because the derivative doesnt exist at sharp ends
#I plot 2 subplots to give perspective on scaling
subplot(211)
title('Two plots of the percent error with different y scaling')
plot(x_new,percent_error)
ylim(0,.01)
ylabel('y')
subplot(212)
plot(x_new,percent_error)
ylim(0,.1)
xlabel('x')
ylabel('y')
show()
```

For the percent error to be very, very small (accurate to 4 sig figs) we need our step size to be .0001, the percent error is down to 4 sig figs if we do this

```python
title('Sine Wave With Noise')
x = linspace(0,2*pi,100)
y = sin(x) + 0.1*random.random(size=x.shape)
plot(x,y)
xlabel('x')
ylabel('y')
show()
```

```python
def fourPtFiniteDiff(x,y):
    """Uses a finite difference method to find the derivative of y with respect to x"""
    dydx = zeros(y.shape,float) #make the dimmension appropriate
    for j in range(2,len(y)-2):
        dydx[j] = (y[j-2]-8*y[j-1]+8*y[j+1]-y[j+2])/(12*(x[1]-x[0]))
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    return dydx
```

The above code lets us use the given formula to numerically figure out the derivative of any
function 'y', the lines of code (6-9) let us figure out the derivative at the begining and end of our interval; we have to do it for 2 points because that's what the formula requires (the y[j+2] bit)

```python
x = arange(0,10.1,0.1)
y = x**5
dydx_ = fourPtFiniteDiff(x,y)
plot(x,dydx_, "g", linewidth=1.5)
plot(x,y_prime, "k--", linewidth=1.5)
xlabel('x')
ylabel('y')
title('The Plot')
show()
```

```python
#Following standard procedure as outlined above to plot percent error:
DIFF=fourPtFiniteDiff(x,y)
Y_PRIME=5*x**4
percent_error_ = abs(Y_PRIME-DIFF)/(Y_PRIME)
plot(x,percent_error_)
ylim(0,.01)
xlabel('x')
ylabel('y')
title("Percent Error vs X")
show()
```

Notice the high percent error at the beginning and end of our interval, this is because this method of numerical differentiation gets shotty at these points
