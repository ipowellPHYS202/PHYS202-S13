```python
from scipy.interpolate import interp1d
```

```python
theta = arange(0,6.14,.1)
newTheta = arange(0,6,.1)
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp = interp1d(theta,sin(theta))
newSin = sinInterp(newTheta)
plot(newTheta,newSin)     
show()
```

```python
C = arange(10,40,1)
R = arange(0,30,1)
T = interp1d(C,R)
new_T=T(C)
plot(new_T,C)
show()
```

```python
R = rand(30)
Resistance=interp1d(C,R,kind='cubic')
Int_Res = Resistance(C)
plot(C,Int_Res)
xlim(15,30)
show()
```
