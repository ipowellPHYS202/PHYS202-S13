Using Boolean Masks For Finding Data
=

```python
theta = arange(0.,2*pi,pi/100)
plot(theta,sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
print mask
plot(theta,sin(theta), theta[mask], sin(theta[mask]))
```

```python
y = sin(theta[mask])
jj = (y == max(y)); print jj
y[jj] = NaN
plot(theta, sin(theta))
plot(theta[mask],y,linewidth=3)
```

Interpolation
=

```python
theta = arange(0.,2*pi+.01,pi/10)
mask = (theta == .5)
```

```python
newTheta = 0.5; newSin = 0.
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp = interp1d(theta,sin(theta))
newSin = sinInterp(newTheta)
print newSin
plot(theta,sin(theta),newTheta,newSin, 'r+', markersize = 20, markeredgewidth=3)     
show()
```

```python
thetaNew = arange(0.,2*pi+.1,pi/30)
sinNew = interp1d(theta, sin(theta))(thetaNew)
plot(theta, sin(theta), thetaNew, sinNew, 'r.')
show()
```

```python
plot(thetaNew, sinNew-sin(thetaNew))
ylabel("difference between interpolated values and exact values")
show()
```

###Let's try a better interpolation: a cubic spline

```python
sinNewC = interp1d(theta, sin(theta), kind='cubic')(thetaNew)
subplot(2,1,1)
xlim(0,2*pi)
plot(thetaNew, sinNew-sin(thetaNew), sinNewC - sin(thetaNew), 'r')
subplot(2,1,2)
plot(thetaNew, sinNewC - sin(thetaNew), 'r')
show()
```

```python
x = arange(0.,10.01,.1)
y = x**5
print x
print y 
```

```python
plot(x,y)
show()
```

```python
dy = diff(y) #y(n+1)-y(n)
#help(diff)
print len(y), len(dy)
dx = diff(x)
dydx = dy/dx
print len(dydx)
```

```python
plot(x,dy/dx)
```

```python
dydx1 = zeros(y.shape,float)
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
```

$y = x^5 , \frac{dy}{dx} = 5x^4$

```python
dydxExact = 5*x**4
```

```python
plot(x,dydx1,x,dydxExact)
show()
```
