```python
#Using the built in product function I can set the entire time frame--giving each year its set of 12 months
Time_Frame = product(range(1901,2001),range(1,13))
#I iterate over all years and months in our time frame *jan 1st 1901-dec 31st 2000*
for (years,months) in Time_Frame:
    #The monthrange function allows me to see if the first of the month [0] is sunday [6]
    if monthrange(years,months)[0]==6:
    #I now append each instance sunday appears as the 1st in the month onto a list "total"
       total = []
       total.append([years,months])
       print len(total[0:])
```

```python
#As you can see, I just get a 1 for each time a sunday appears as the first.  I tried a bunch of
#different stuff, but I couldn't get the previous code to print the exact number, so here I tried
#something different:
#setting the time frame again:
Time_Frame = product(range(1901,2001),range(1,13))
#condensing all of the code in the previous cell to one line
#I now print the length of the new list with only the appropriate sundays being members
#I use the familiar monthrange function to do so:
print len([(years,months) for (years, months) in Time_Frame if monthrange(years, months)[0] == 6])
```
