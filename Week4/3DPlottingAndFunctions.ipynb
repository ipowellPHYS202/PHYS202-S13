Electric Dipole Potential
=

$V(x,y) = \frac{kq}{\sqrt{x^2+(y-d)^2}} - \frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
#make an x-y grid to cover plane -2<x,y<2
dx = .06
dy = .06
x = arange(-2., 2., dx)
y = arange(-2.,2.,dy)
print x.ndim
print len(x)

#meshgrid creates a 2d array essentially 
X,Y = meshgrid(x,y)

print X
print Y
```

```python
k =1; q=1; d=1
V=(k*q/(X**2+(Y-d)**2)**(1/2.))-(k*q/(X**2+(Y+d)**2)**(1/2.))
```

```python
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection ='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=1.0, cmap=copper)
```

```python
#Representing 3d data with countors
levels=array([8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8])
#this "levels" array tells python what contours to plot
contour(X,Y,V,levels,linewidths=2,cmap=cm.Dark2)
#cmap tells python what colors to use when plotting lines
show()
```

###Ex 2: Determine the electric potential for a quadropole consisting of q=-1.0nC at both (x=+-2, y=0) and q = 1nC at both (x=0,y=+-2). 

```python
###Make a surface plot and a contour plot on the range -4<x,y<4 using 101 bins in x and y.
```

```python
###What is the value of the function at V(20,30)?
```

```python
dx = (4./101.)
dy = (4./101.)
x = arange(-2., 2., dx)
y = arange(-2.,2.,dy)

X,Y = meshgrid(x,y)
```

```python
k = 8.987551787*10**9;q=1*10**-9
#k is in units of Nm^2/C^2, q is in unit of C^2, distance is in m

#Adding the scalar potentials~
V = (-k*q/(X**2+(Y-d)**2)**1/2.)-(k*q/(X**2+(Y+d)**2)**1/2.)+(k*q/((X-d)**2+Y**2)**1/2.)+(k*q/((X+d)**2+Y**2)**1/2.)
```

```python
print V
```

```python
levels = array([16,8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8-16])
contour(X,Y,V,levels,linewidths=2,cmap=cm.autumn)
xlim(-2.,2.)
ylim(-2.,2.)
show()
```
