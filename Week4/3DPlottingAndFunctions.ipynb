Electric Dipole Potential
=

$V(x,y) = \frac{kq}{\sqrt{x^2+(y-d)^2}} - \frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
#make an x-y grid to cover plane -2<x,y<2
dx = .06
dy = .06
x = arange(-2., 2., dx)
y = arange(-2.,2.,dy)
print x.ndim
print len(x)

#meshgrid creates a 2d array essentially 
X,Y = meshgrid(x,y)

print X
print Y
```

```python
#We simply set k,d,q=1 for simplicity
k =1; q=1; d=1
V_=(k*q/(X**2+(Y-d)**2)**(1/2.))-(k*q/(X**2+(Y+d)**2)**(1/2.))
```

```python
print V_
```

```python
#I import the module then plot the dipole potential
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection ='3d')
xlim(-1.5, 1.5)
ylim(-1.5,1.5)
p = ax.plot_surface(X, Y, V_, rstride=1, cstride=1, linewidth=1.0, cmap=cm.jet)
```

```python
#Representing 3d data with countors
levels=array([8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8])
#this "levels" array tells python what contours to plot
contour(X,Y,V_,levels,linewidths=2,cmap=cm.Dark2)
#cmap tells python what colors to use when plotting lines
show()
```

###Ex 2: Determine the electric potential for a quadropole consisting of q=-1.0nC at both (x=+-2, y=0) and q = 1nC at both (x=0,y=+-2). 

```python
###Make a surface plot and a contour plot on the range -4<x,y<4 using 101 bins in x and y.
```

```python
#I make a mesh grind with the appropriate values
dx = (8./101.)
dy = (8./101.)
x = arange(-4., 4., dx)
y = arange(-4.,4.,dy)

X,Y = meshgrid(x,y)
```

```python
k = 8.987551787*10**9;q=1*10**-9;d=2
#k is in units of Nm^2/C^2, q is in unit of C^2, distance is in m

#Adding the scalar potentials~
V = (-k*q/(X**2+(Y-d)**2)**1/2.)-(k*q/(X**2+(Y+d)**2)**1/2.)+(k*q/((X-d)**2+Y**2)**1/2.)+(k*q/((X+d)**2+Y**2)**1/2.)
```

```python
print V
```

```python
levels = array([16,8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8-16])
contour(X,Y,V,levels,linewidths=2,cmap=cm.autumn)
xlim(-4,4)
ylim(-4,4)
show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection ='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=1.0, cmap=cm.jet)
```

```python
###I define my function pointPotential.  It can find the value of potential anywhere from any charge
k = 8.987551787*10**9
def pointPotential(X,Y,q,posx,posy):
    return (k*q/((X-posx)**2+(Y-posy)**2)**(1/2.))
```

```python
x = 1.; y = 1.
V11 = pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
x,y = meshgrid(arange(-2.,2.,0.06),arange(-2.,2.,0.06))
Vxy = pointPotential(x,y,1e-9,0.,0.)
print Vxy
```

```python
#I add a few arguments to dipole so that it can take any pair of charges.
#Note that I'm not simply subtracting a potential, I let the value of q dictate that.
def dipolePotential(X,Y,q1,q2,d1,d2):
    return (pointPotential(X,Y,q1,-d1/2,-d2/2)+pointPotential(X,Y, q2, d1/2,d2/2))
```

```python
###I create a meshgrid for plotting purposes
dx = .06
dy = .06
x = arange(-2., 2., dx)
y = arange(-2.,2.,dy)
print x.ndim
print len(x)

X,Y = meshgrid(x,y)
```

```python
#Plotting the dipole with arbitrart d and q values:
fig = figure()
ax = fig.gca(projection ='3d')
p = ax.plot_surface(X, Y, dipolePotential(X,Y,1*10**-9,-1*10**-9,2,0), rstride=1, cstride=1, linewidth=1.0, cmap=cm.jet)
```

```python
#Plotting the quadropole with arbitrary d and q values
fig = figure()
ax = fig.gca(projection ='3d')
p = ax.plot_surface(X, Y, dipolePotential(X,Y,1*10**-9,1*10**-9,2,0)+dipolePotential(X,Y,-1*10**-9,-1*10**-9,0,2), rstride=1, cstride=1, linewidth=1.0, cmap=cm.jet)
```
