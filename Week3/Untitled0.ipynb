```python
data = loadtxt("sunspots.txt", float)
x = data[:1001:,0]
y = data[:1001:,1]
```

```python
plot(x, y, label ="Number of Observed Sunspots")
#Defining the rolling average function "avg(a).  Trivial code.
def avg(a):
    sum_ = 0
    number = 1
    for i in a:
        sum_ += i
        number += 1
    return sum_/number
#Making the rolling average function relevant
a = range(1001)
def rolling_avg(a):
    for i in a:
        a[i] = avg(y[:i])
    return a
#plotting the rolling average
plot(x, rolling_avg(a), label = "running average")
xlabel("time (years)")
ylabel("number of sunspots")
title("Observed Sunspots versus Time")
legend()
#Sizing our x-axis to be historically relevant
xticks(arange(0,1000, 1000/6), [1749, 1762, 1776, 1790, 1804, 1818, 1832])
grid(True)
show()
```
