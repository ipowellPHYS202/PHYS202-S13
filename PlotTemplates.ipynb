###Publication ready plot template: Simple graphing template for matplotlib functions and data plots

Line Plots

```python
y = arange(-5, 5, 0.001) 
E = cosh(y) 
p = sinh(y) 
#Here we slice up y in an array so that we can plot many values in the range -5 to 5.
#We define E and p as the above hyperbolic functions already installed in python.
```

```python
#First, we plot our functions E and p vs y.
plot(y, E, 'r:', linewidth = 1.5, label = "$Function (I)$")
plot(y, p, 'g--', linewidth = 2.0, label = "$Function (II)$")
ax = gca() #getting axis info
ax.spines['right'].set_color('none') #setting color to none!
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axhline(y=0, color = "black")
axvline(x=0, color = "black")
title("Graph of Function (I) and Function (II)")
legend(loc="lower right")
xlabel("x")
ylabel("y")
show()
```

Sub Plots


```python
subplot(1,2,1) #Here I created a subplot with two colums and a row (hence 1,2,graph 1)
plot(y, E, 'r:', linewidth = 2.5, label = "$Function (I)$") #plotting Function (I)
xlabel("x")
ylabel("y")
title("Function (I)")
ax = gca() 
ax.spines['right'].set_color('none') #setting color of boarders to none!
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axhline(y=40, color = "black") #creating axes through middle of graph
axvline(x=0, color = "black")
legend()

subplot(1,2,2) #The same process follows for the the second function
plot(y, p, 'b--', linewidth = 3, label = "$Function (II)$") 
xlabel("x")
ylabel("y")
title("Function (II)")
ax = gca() 
ax.spines['right'].set_color('none') 
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axhline(y=0, color = "black")
axvline(x=0, color = "black")
legend()
show()

```

```python
#We simply plot the scatter instead of function and follow the same steps as before
scatter(X,Y, s = 1.5, label = "$Scatter plot A$")
ax = gca() 
ax.spines['right'].set_color('none') 
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axhline(y=0, color = "black")
axvline(x=0, color = "black")
title("Scatter Plot X&Y")
legend(loc="lower right")
xlabel("X")
ylabel("Y")
show()
```
